```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Utilities

```python
def plot_all_scores(df, name):
    df_se = df['MSE_all'].sort_values(ascending = False)
    df_ae = df['MAE_all'].sort_values(ascending = False)
    df_r = df['R2_all'].sort_values(ascending = True)

    fig, ax = plt.subplots(1,3, figsize = (13,5), tight_layout = True)
    df_se.plot(ax = ax[0], style='s-', color = 'purple')
    ax[0].grid()
    ax[0].set_title('MSE scores ')
    ax[0].set_xticks(range(len(df.index)))
    ax[0].set_xticklabels(df_se.index)
    ax[0].tick_params(axis='x', labelrotation=90)

    df_ae.plot(ax = ax[1],style='s-', color = 'green')
    ax[1].grid()
    ax[1].set_title('MAE scores')
    ax[1].tick_params(axis='x', labelrotation=90)
    ax[1].set_xticks(range(len(df.index)))
    ax[1].set_xticklabels(df_ae.index)

    df_r.plot(ax = ax[2],style='s-', color = 'crimson')
    ax[2].grid()
    ax[2].set_title('$R^2$ scores ')
    ax[2].tick_params(axis='x', labelrotation=90)
    ax[2].set_xticks(range(len(df.index)))
    ax[2].set_xticklabels(df_r.index)

    plt.savefig(f'{name}_test_accuracies_all')
```

```python
def plot_mse_score(df_mse, name):
    df8_mse11 = df_mse['MSE_h11'].sort_values(ascending = False)
    df8_mse21 = df_mse['MSE_h21'].sort_values(ascending = False)
    df8_mse31 = df_mse['MSE_h31'].sort_values(ascending = False)
    df8_mse22 = df_mse['MSE_h22'].sort_values(ascending = False)

    fig, ax = plt.subplots(2,2, figsize = (10,10), tight_layout = True)
    df8_mse11.plot(ax = ax[0,0], style='s-', color = 'purple')
    ax[0,0].grid()
    ax[0,0].set_title('MSE scores for $h^{(1,1)}$')
    ax[0,0].set_xticks(range(len(df_mse.index)))
    ax[0,0].set_xticklabels(df8_mse11.index)
    ax[0,0].tick_params(axis='x', labelrotation=90)

    df8_mse21.plot(ax = ax[0,1],style='s-', color = 'purple')
    ax[0,1].grid()
    ax[0,1].set_title('MSE scores for $h^{(2,1)}$')
    ax[0,1].tick_params(axis='x', labelrotation=90)
    ax[0,1].set_xticks(range(len(df_mse.index)))
    ax[0,1].set_xticklabels(df8_mse21.index)

    df8_mse31.plot(ax = ax[1,0],style='s-', color = 'purple')
    ax[1,0].grid()
    ax[1,0].set_title('MSE scores for $h^{(3,1)}$')
    ax[1,0].tick_params(axis='x', labelrotation=90)
    ax[1,0].set_xticks(range(len(df_mse.index)))
    ax[1,0].set_xticklabels(df8_mse31.index)

    df8_mse22.plot(ax = ax[1,1],style='s-', color = 'purple')
    ax[1,1].grid()
    ax[1,1].set_title('MSE scores for $h^{(2,2)}$')
    ax[1,1].tick_params(axis='x', labelrotation=90)
    ax[1,1].set_xticks(range(len(df_mse.index)))
    ax[1,1].set_xticklabels(df8_mse22.index)

    plt.savefig(f'{name}_test_accuracies_all')
```

```python
def plot_mae_score(df_mae, name):
    df8_mae11 = df_mae['MAE_h11'].sort_values(ascending = False)
    df8_mae21 = df_mae['MAE_h21'].sort_values(ascending = False)
    df8_mae31 = df_mae['MAE_h31'].sort_values(ascending = False)
    df8_mae22 = df_mae['MAE_h22'].sort_values(ascending = False)

    fig, ax = plt.subplots(2,2, figsize = (10,10), tight_layout = True)
    df8_mae11.plot(ax = ax[0,0], style='s-', color = 'green')
    ax[0,0].grid()
    ax[0,0].set_title('MAE scores for $h^{(1,1)}$')
    ax[0,0].set_xticks(range(len(df_mae.index)))
    ax[0,0].set_xticklabels(df8_mae11.index)
    ax[0,0].tick_params(axis='x', labelrotation=90)

    df8_mae21.plot(ax = ax[0,1],style='s-', color = 'green')
    ax[0,1].grid()
    ax[0,1].set_title('MAE scores for $h^{(2,1)}$')
    ax[0,1].tick_params(axis='x', labelrotation=90)
    ax[0,1].set_xticks(range(len(df_mae.index)))
    ax[0,1].set_xticklabels(df8_mae21.index)

    df8_mae31.plot(ax = ax[1,0],style='s-', color = 'green')
    ax[1,0].grid()
    ax[1,0].set_title('MAE scores for $h^{(3,1)}$')
    ax[1,0].tick_params(axis='x', labelrotation=90)
    ax[1,0].set_xticks(range(len(df_mae.index)))
    ax[1,0].set_xticklabels(df8_mae31.index)

    df8_mae22.plot(ax = ax[1,1],style='s-', color = 'green')
    ax[1,1].grid()
    ax[1,1].set_title('MAE scores for $h^{(2,2)}$')
    ax[1,1].tick_params(axis='x', labelrotation=90)
    ax[1,1].set_xticks(range(len(df_mae.index)))
    ax[1,1].set_xticklabels(df8_mae22.index)

    plt.savefig(f'{name}_test_accuracies_all')
```

```python
def plot_r2_score(df_r2, name):
    df8_r11 = df_r2['r2_h11'].sort_values(ascending = True)
    df8_r21 = df_r2['r2_h21'].sort_values(ascending = True)
    df8_r31 = df_r2['r2_h31'].sort_values(ascending = True)
    df8_r22 = df_r2['r2_h22'].sort_values(ascending = True)

    fig, ax = plt.subplots(2,2, figsize = (10,10), tight_layout = True)
    df8_r11.plot(ax = ax[0,0], style='s-', color = 'crimson')
    ax[0,0].grid()
    ax[0,0].set_title('$R^2$ scores for $h^{(1,1)}$')
    ax[0,0].set_xticks(range(len(df_r2.index)))
    ax[0,0].set_xticklabels(df8_r11.index)
    ax[0,0].tick_params(axis='x', labelrotation=90)

    df8_r21.plot(ax = ax[0,1],style='s-', color = 'crimson')
    ax[0,1].grid()
    ax[0,1].set_title('$R^2$ scores for $h^{(2,1)}$')
    ax[0,1].tick_params(axis='x', labelrotation=90)
    ax[0,1].set_xticks(range(len(df_r2.index)))
    ax[0,1].set_xticklabels(df8_r21.index)

    df8_r31.plot(ax = ax[1,0],style='s-', color = 'crimson')
    ax[1,0].grid()
    ax[1,0].set_title('$R^2$ scores for $h^{(3,1)}$')
    ax[1,0].tick_params(axis='x', labelrotation=90)
    ax[1,0].set_xticks(range(len(df_r2.index)))
    ax[1,0].set_xticklabels(df8_r31.index)

    df8_r22.plot(ax = ax[1,1],style='s-', color = 'crimson')
    ax[1,1].grid()
    ax[1,1].set_title('$R^2$ scores for $h^{(2,2)}$')
    ax[1,1].tick_params(axis='x', labelrotation=90)
    ax[1,1].set_xticks(range(len(df_r2.index)))
    ax[1,1].set_xticklabels(df8_r22.index)

    plt.savefig(f'{name}_test_accuracies_all')
```

# Check last nonzero values

```python
loss_dict_resnet_gru_256 = loss_dict_resnet_gru_256[:500]
```

```python
#write a function to check for the last nonzero value
def check_last_nonzero(df, model_name):
    for i in range(len(df)):
        if df['train_loss'][i] == 0 and df['train_loss'][i+1] == 0:
            return model_name, i-1, df['train_loss'][i-1], df['val_loss'][i-1]
```

```python
check_last_nonzero(loss_dict_resnet_gru_256, 'ResNet-GRU-256')
```
